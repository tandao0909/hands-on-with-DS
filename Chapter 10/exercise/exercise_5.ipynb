Here, I will only talk about single-variable activation functions:
- The step function (the heaviside function)
- The ReLU function
- The sigmoid function
- The tanh function

```python
from scipy.special import expit as sigmoid
import numpy as np


def relu(z):
    return np.maximum(z, 0)
```

```python
import matplotlib.pyplot as plt

max_z = 5
z = np.linspace(-max_z, max_z, 100)

plt.figure(figsize=(8, 3))
plt.plot([-max_z, 0], [0, 0], "r-", label="Heaviside")
plt.plot([0, 0], [0, 1], "r-")
plt.plot([0, max_z], [1, 1], "r-")
plt.plot(z, relu(z), "m-.", label="ReLU")
plt.plot(z, sigmoid(z), "g--", label="Sigmoid")
plt.plot(z, np.tanh(z), "b-", label="Tanh")
plt.legend(loc="lower right")
plt.xticks(np.arange(-4, 5, 1))
plt.axis([-max_z, max_z, -1.5, 2.5])
plt.grid()
```
